# Point defect in 2D Photonic Crystal

## Cavity in 2D PC

```python
import meep as mp
from meep import mpb
import numpy as np

import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 16})
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": ["Helvetica"]})
# for Palatino and other serif fonts use:
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Palatino"],
})
```

```python
resolution = 64
```

```python
a = 1 # arbitrary units
sc = 5 * a  

geometry_lattice = mp.Lattice(size=mp.Vector3(5*a,5*a))
geometry = [mp.Cylinder(0.2, material=mp.Medium(epsilon=12))]
geometry = mp.geometric_objects_lattice_duplicates(geometry_lattice, geometry)
geometry.append(mp.Cylinder(0.2, material=mp.air))
```

```python
sim = mp.Simulation(resolution=resolution,
                    cell_size= mp.Vector3(sc,sc,0),
                    geometry=geometry)
sim.init_sim()
eps_data = sim.get_epsilon()
```

```python
fig1, ax1 = plt.subplots(figsize=(10,6))
ax1.set_title('Supercell of the 2D photonic crystal')
pos1 = ax1.imshow(eps_data, interpolation='spline36', cmap='binary')
ax1.axis('off')
fig1.colorbar(pos1, ax=ax1)
```

So as we can see, we have created the point defect!

```python
num_bands = 40
ms = mpb.ModeSolver(geometry_lattice=geometry_lattice,
                    geometry=geometry,
                    k_points=[mp.Vector3(0.5, 0.5)],
                    resolution=resolution,
                    num_bands=num_bands);
```

```python
ms.run();
```

```python
dfields = []

def get_dfields(ms, band):
    dfields.append(ms.get_dfield(band, bloch_phase=True))

ms.run_tm(mpb.output_at_kpoint(mp.Vector3(0.5, 0.5), mpb.fix_dfield_phase,
          get_dfields))

md = mpb.MPBData(rectify=True, resolution=resolution, periods=1)

converted = []

for f in dfields:
    # Get just the z component of the dfields
    f = f[..., 0, 2]
    converted.append(md.convert(f))
```

```python
fig2, ax2 = plt.subplots(1,2,figsize=(16,6))
ax2[0].set_title('$D_z$ fields at $M$ for point defect')
ax2[1].set_title('$D_z$ fields at $M$ for point defect')
ax2[0].contour(eps_data.T, cmap='binary')
ax2[1].contour(eps_data.T, cmap='binary')
posd1 = ax2[0].imshow(np.real(converted[24]).T, interpolation='spline36', cmap='plasma')
posd2 = ax2[1].imshow(np.real(converted[25]).T, interpolation='spline36', cmap='plasma')
ax2[0].set_xticks([])
ax2[0].set_xticklabels([]);
ax2[0].set_yticks([])
ax2[0].set_yticklabels([]);
ax2[1].set_xticks([])
ax2[1].set_xticklabels([]);
ax2[1].set_yticks([])
ax2[1].set_yticklabels([]);
fig2.colorbar(posd1,ax=ax2[0])
fig2.colorbar(posd2,ax=ax2[1])
```

On the left, localized mode (corresponding to a monopole),  in contrast to a guided mode on the right!

##  Higher order modes

```python
a = 1 # arbitrary units
sc = 5 * a  

geometry_lattice = mp.Lattice(size=mp.Vector3(5*a,5*a))
geometry = [mp.Cylinder(0.2, material=mp.Medium(epsilon=8.9))]
geometry = mp.geometric_objects_lattice_duplicates(geometry_lattice, geometry)
geometry.append(mp.Cylinder(0.55, material=mp.Medium(epsilon=8.9)))
```

```python
resolution = 64

sim = mp.Simulation(resolution=resolution,
                    cell_size= mp.Vector3(sc,sc,0),
                    geometry=geometry)
sim.init_sim()
eps_data = sim.get_epsilon()
```

```python
fig1, ax1 = plt.subplots(figsize=(10,6))
ax1.set_title('Supercell of the 2D photonic crystal')
pos1 = ax1.imshow(eps_data, interpolation='spline36', cmap='binary')
ax1.axis('off')
fig1.colorbar(pos1, ax=ax1)
```

```python
num_bands = 40
ms = mpb.ModeSolver(geometry_lattice=geometry_lattice,
                    geometry=geometry,
                    k_points=[mp.Vector3(0.5, 0.5)],
                    resolution=resolution,
                    num_bands=num_bands);
```

```python
ms.run();
```

```python
dfields = []

def get_dfields(ms, band):
    dfields.append(ms.get_dfield(band, bloch_phase=True))

ms.run_tm(mpb.output_at_kpoint(mp.Vector3(0.5, 0.5), mpb.fix_dfield_phase,
          get_dfields))

md = mpb.MPBData(rectify=True, resolution=resolution, periods=1)

converted = []

for f in dfields:
    # Get just the z component of the dfields
    f = f[..., 0, 2]
    converted.append(md.convert(f))
```

```python
fig2, ax2 = plt.subplots(1,2,figsize=(16,6))
ax2[0].set_title('$D_z$ fields at $M$ for point defect')
ax2[1].set_title('$D_z$ fields at $M$ for point defect')
ax2[0].contour(eps_data.T, cmap='binary')
ax2[1].contour(eps_data.T, cmap='binary')
posd1 = ax2[0].imshow(np.real(converted[27]).T, interpolation='spline36', cmap='plasma')
posd2 = ax2[1].imshow(np.real(converted[29]).T, interpolation='spline36', cmap='plasma')
ax2[0].set_xticks([])
ax2[0].set_xticklabels([]);
ax2[0].set_yticks([])
ax2[0].set_yticklabels([]);
ax2[1].set_xticks([])
ax2[1].set_xticklabels([]);
ax2[1].set_yticks([])
ax2[1].set_yticklabels([]);
fig2.colorbar(posd1,ax=ax2[0])
fig2.colorbar(posd2,ax=ax2[1])
```

Here we can see a quadrupole mode and a monopole-2 mode pulled into the gap!
