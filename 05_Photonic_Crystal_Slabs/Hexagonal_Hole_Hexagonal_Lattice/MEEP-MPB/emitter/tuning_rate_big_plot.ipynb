```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

plt.style.use("science")
matplotlib.rcParams.update({"font.size": 22})
```

```python
fig, axi = plt.subplots(3,3, figsize=(20,16))

pos_0_nm_x_tensor = np.loadtxt('data/0.0_nm_x_axis_green_tensor.txt')
pos_100_nm_x_tensor = np.loadtxt('data/100_nm_x_axis_green_tensor.txt')
pos_200_nm_x_tensor = np.loadtxt('data/200_nm_x_axis_green_tensor.txt')

i = 0

all_arrays = np.array([pos_0_nm_x_tensor, pos_100_nm_x_tensor, pos_200_nm_x_tensor])

colors = [['skyblue', 'royalblue', 'darkblue', 'royalblue', 'skyblue'], ['lightcoral', 'indianred', 'darkred', 'indianred', 'lightcoral'], ['palegreen','forestgreen','darkgreen', 'forestgreen', 'palegreen']]

color_zero = ['b','r','g']

for ax in axi[0]:

    ax.plot(np.linspace(300,400,500),all_arrays[0][i], label='x = 0 nm', color=colors[i][0], lw=2)
    ax.plot(np.linspace(300,400,500),all_arrays[1][i], label='x = $\pm$100 nm', color=colors[i][1], lw=2)
    ax.plot(np.linspace(300,400,500),all_arrays[2][i], label='x = $\pm$200 nm', color=colors[i][2], lw=2)

    #ax.plot(np.linspace(300,400,500),np.ones(500), linestyle='dashed', color=color_zero[i], alpha=0.7)
    ax.set_yscale('log')
    ax.set_ylim([0.1,10])
    #ax.set_xlabel('$\\nu$ (THz)')
    ax.legend(frameon=True, fontsize=16)
    
    i += 1

pos_0_nm_y_tensor = np.loadtxt('data/0.0_nm_y_axis_green_tensor.txt')
pos_100_nm_y_tensor = np.loadtxt('data/100_nm_y_axis_green_tensor.txt')
pos_200_nm_y_tensor = np.loadtxt('data/200_nm_y_axis_green_tensor.txt')

all_arrays = np.array([pos_0_nm_y_tensor, pos_100_nm_y_tensor, pos_200_nm_y_tensor])
i = 0
for ax in axi[1]:

    ax.plot(np.linspace(300,400,500),all_arrays[0][i], label='y = 0 nm', color=colors[i][0], lw=2)
    ax.plot(np.linspace(300,400,500),all_arrays[1][i], label='y = $\pm$100 nm', color=colors[i][1], lw=2)
    ax.plot(np.linspace(300,400,500),all_arrays[2][i], label='y = $\pm$200 nm', color=colors[i][2], lw=2)

    #ax.plot(np.linspace(300,400,500),np.ones(500), linestyle='dashed', color=color_zero[i], alpha=0.7)
    ax.set_yscale('log')
    ax.set_ylim([0.1,10])
    #ax.set_xlabel('$\\nu$ (THz)')
    ax.legend(frameon=True, fontsize=16)
    
    i += 1
    
pos_0_nm_z_tensor = np.loadtxt('data/0.0_nm_z_axis_green_tensor.txt')
pos_100_nm_z_tensor = np.loadtxt('data/100_nm_z_axis_green_tensor.txt')
pos_200_nm_z_tensor = np.loadtxt('data/200_nm_z_axis_green_tensor.txt')

all_arrays = np.array([pos_0_nm_z_tensor, pos_100_nm_z_tensor, pos_200_nm_z_tensor])
i = 0
for ax in axi[2]:

    ax.plot(np.linspace(300,400,500),all_arrays[0][i], label='z = 0 nm', color=colors[i][0], lw=2)
    ax.plot(np.linspace(300,400,500),all_arrays[1][i], label='z = $\pm$100 nm', color=colors[i][1], lw=2)
    ax.plot(np.linspace(300,400,500),all_arrays[2][i], label='z = $\pm$200 nm', color=colors[i][2], lw=2)

    #ax.plot(np.linspace(300,400,500),np.ones(500), linestyle='dashed', color=color_zero[i], alpha=0.7)
    ax.set_yscale('log')
    ax.set_ylim([0.1,10])
    ax.set_xlabel('$\\nu$ (THz)')
    ax.legend(frameon=True, fontsize=18)
    
    i += 1
    
axi[0,0].set_ylabel('PF$\equiv\Gamma_{\\text{tot}}/\Gamma_0$')
axi[1,0].set_ylabel('PF$\equiv\Gamma_{\\text{tot}}/\Gamma_0$')
axi[2,0].set_ylabel('PF$\equiv\Gamma_{\\text{tot}}/\Gamma_0$')
```
