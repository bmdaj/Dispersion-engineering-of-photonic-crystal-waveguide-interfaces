# Emission rates

In this section we will calculate create plots for the Purcell Factor and $\Gamma_{2D}/\Gamma$ for the TE dielectric band and different dipole orientations.

## Purcell Factor

First of all, we load the files where we placed the atom at the center of the holes in the PhC:

```python
G_function = np.loadtxt('data/0.0_nm_x_axis_green_tensor.txt')

G_x_function = G_function [0,:]
G_y_function = G_function [1,:]
G_z_function = G_function [2,:]
```

```python
freq_range = np.linspace(300,400,500) #THz
```

```python
_, Gamma_1D_over_prime_x = np.loadtxt('data/Emission_data_X.txt')
_, Gamma_1D_over_prime_y = np.loadtxt('data/Emission_data_Y.txt')
```

```python
fig, ax = plt.subplots(1,2,figsize=(18,8))
ax0, ax1 = ax

ax0.plot(freq_range, G_x_function , color='red', label='$G_{xx}$')
ax0.plot(freq_range, G_y_function, color='blue', label='$G_{yy}$')
ax0.plot(freq_range, G_z_function, color='green', label='$G_{zz}$')

ax0.set_ylabel('PF = $\\Gamma/\\Gamma_0$')
ax0.set_xlabel('$\\nu$ (THz)')
ax0.set_yscale('log')
ax0.legend()

ax1.plot(freq_range, Gamma_1D_over_prime_x,  color='red', label='X orientation')
ax1.plot(freq_range, Gamma_1D_over_prime_y,  color='blue', label='Y orientation')


ax1.set_ylabel('$\\Gamma_{2D}/\\Gamma\'$')
ax1.set_xlabel('$\\nu$ (THz)')
ax1.set_yscale('log')
ax1.legend()
```
