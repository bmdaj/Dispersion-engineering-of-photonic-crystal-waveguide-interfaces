```python
import meep as mp 
import numpy as np
import matplotlib.pyplot as plt
from hexalattice.hexalattice import *

plt.rcParams.update({'font.size': 16})
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": ["Helvetica"]})
# for Palatino and other serif fonts use:
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Palatino"],
})
```

```python
pos_0_nm_x_tensor = np.loadtxt('data/0.0_nm_x_axis_green_tensor.txt')
pos_100_nm_x_tensor = np.loadtxt('data/100_nm_x_axis_green_tensor.txt')
pos_200_nm_x_tensor = np.loadtxt('data/200_nm_x_axis_green_tensor.txt')
```

```python
fig, axes = plt.subplots(3, figsize=(16,35))

i = 0

all_arrays = np.array([pos_0_nm_x_tensor, pos_100_nm_x_tensor, pos_200_nm_x_tensor])

for ax in axes:
        ax.plot(np.linspace(300,400,500),all_arrays[i][0], label='X orientation', color='blue')
        ax.plot(np.linspace(300,400,500),all_arrays[i][1], label='Y orientation', color='red')
        ax.plot(np.linspace(300,400,500),all_arrays[i][2], label='Z orientation', color='green')


        ax.plot(np.linspace(300,400,500),np.ones(500), linestyle='dashed', color='black', alpha=0.7)
        ax.set_yscale('log')
        ax.set_ylim([0.1,10])
        ax.set_ylabel('Im$(\mathbf{\\bar{G}}_{ii}(\\nu))$/Im$(\mathbf{\\bar{G}}^{0}_{ii}(\\nu))$')
        ax.set_xlabel('$\\nu$ (THz)')
        ax.legend()
    
        i += 1
    
axes[0].set_title('Rate of spontaneous emission for $x=0$ nm')
axes[1].set_title('Rate of spontaneous emission for $x=100$ nm')
axes[2].set_title('Rate of spontaneous emission for $x=200$ nm')
```

```python
fig.savefig('results/green_tensor_X_direction.pdf')
```

Now we plot the change for each dipole direction:

```python
fig, axes = plt.subplots(3,figsize=(12,26))

i = 0

all_arrays = np.array([pos_0_nm_x_tensor, pos_100_nm_x_tensor, pos_200_nm_x_tensor])

colors = [['skyblue', 'royalblue', 'darkblue', 'royalblue', 'skyblue'], ['lightcoral', 'indianred', 'darkred', 'indianred', 'lightcoral'], ['palegreen','forestgreen','darkgreen', 'forestgreen', 'palegreen']]

color_zero = ['b','r','g']

for ax in axes:

    ax.plot(np.linspace(300,400,500),all_arrays[0][i], label='x = 0 nm', color=colors[i][0])
    ax.plot(np.linspace(300,400,500),all_arrays[1][i], label='x = 100 nm', color=colors[i][1])
    ax.plot(np.linspace(300,400,500),all_arrays[2][i], label='x = 200 nm', color=colors[i][2])

    ax.plot(np.linspace(300,400,500),np.ones(500), linestyle='dashed', color=color_zero[i], alpha=0.7)
    ax.set_yscale('log')
    ax.set_ylim([0.1,10])
    ax.set_xlabel('$\\nu$ (THz)')
    ax.legend()
    
    i += 1

    
axes[0].set_ylabel('Im$(\mathbf{\\bar{G}}_{xx}(\\nu))$/Im$(\mathbf{\\bar{G}}^{0}_{xx}(\\nu))$')
axes[1].set_ylabel('Im$(\mathbf{\\bar{G}}_{yy}(\\nu))$/Im$(\mathbf{\\bar{G}}^{0}_{yy}(\\nu))$')
axes[2].set_ylabel('Im$(\mathbf{\\bar{G}}_{zz}(\\nu))$/Im$(\mathbf{\\bar{G}}^{0}_{zz}(\\nu))$')
```

```python
fig.savefig('results/green_function_X_direction.pdf')
```
