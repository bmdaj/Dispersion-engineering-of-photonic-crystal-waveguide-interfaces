##  Defining the sinusoidal dielectric blocks

We define some boolean values for executing purposes:

```python
plot = False
simulate = False
save = False
```

First, we define some parameters that describe the geometry:

```python
a = 335e-9  # in nm
w = 335e-9 / a
t = 200e-9 / a
r = 0.5 * 116e-9 / a
g = 250e-9 / a
index = 1.9935  # refractive index of Si3N4
az = 15

# The cell dimensions
sx = a / a
sy = 15     #size of cell in y direction (perpendicular to wvg.)
sz = 15     # size of cell in z direction (perpendicular to wvg.)ax = 3.7                                    #  e-7 meters


cell = mp.Vector3(sx, sy, sz)
geometry_lattice = mp.Lattice(size=mp.Vector3(sx,sy, sz))


resolution = 4
```

We define the parameters for the APCW:

```python
geometry = [mp.Block(center=mp.Vector3(0,-0.5*(g+w),0), 
            size=mp.Vector3(mp.inf,w,t), 
            material=mp.Medium(index = 2))]                    # First Si3N4 waveguide block

geometry.append(mp.Block(center=mp.Vector3(0,0.5*(g+w),0), 
            size=mp.Vector3(mp.inf,w,t), 
            material=mp.Medium(index = 2)))  

for i in range(int(sx)):
    geometry.append(mp.Cylinder(center=mp.Vector3(int(0.5*i),0.5*(g+w),0),
                               height = t,
                               radius = r,
                               material = mp.Medium(index = 1)))
    geometry.append(mp.Cylinder(center=mp.Vector3(int(0.5*i),-0.5*(g+w),0),
                               height = t,
                               radius = r,
                               material = mp.Medium(index = 1)))
    geometry.append(mp.Cylinder(center=mp.Vector3(-int(0.5*i),-0.5*(g+w),0),
                               height = t,
                               radius = r,
                               material = mp.Medium(index = 1)))
    geometry.append(mp.Cylinder(center=mp.Vector3(-int(0.5*i),0.5*(g+w),0),
                               height = t,
                               radius = r,
                               material = mp.Medium(index = 1)))
```

## Band diagram calculation:

```python
import os
path = os.path.abspath(os.getcwd())
print(path)
```

```python
f = np.genfromtxt(path+"/bands_zeven.dat", delimiter=",")[:,1:]
```

```python
from scipy.interpolate import make_interp_spline, BSpline

fig, ax = plt.subplots(figsize=(14,10));

light_speed = 3e8

num_bands = 5

xlims = [0.6,1]
ylims = [280,440]

for i in range(1,num_bands):
    xnew = np.linspace((f[1:,1]*2).min(), (f[1:,1]*2).max(), 300) 
    spl1 = make_interp_spline(f[1:,1]*2, f[1:,-i]*light_speed/(a*10**12), k=3)  # type: BSpline
    mode_smooth = spl1(xnew)
    if i ==1:
        ax.plot(xnew, mode_smooth, c='red', linewidth=2, label='Propagation modes')
    elif i ==4:
         ax.plot(xnew, mode_smooth, c='blue',linewidth=2, label="Probe band") 
    else:
        ax.plot(xnew, mode_smooth, c='red',linewidth=2) 
    

ax.plot(xnew, np.max(mode_smooth)*np.ones(len(xnew)), linestyle='dotted', c='blue', linewidth=2, label="Probe frequency")#: "+str(round(np.max(mode_smooth),2))+" THz"") 
           
ax.plot(np.linspace(0,1,100),np.linspace(0,1,100)*light_speed/(2*a*10**12), linewidth=2, c='black')
ax.fill_between(np.linspace(0,1,100), np.linspace(0,1,100)*light_speed/(2*a*10**12), 1*np.ones(100)*light_speed/(2*a*10**12),color='black', label="Light cone", alpha=0.65) 

#ax.set_title('Dispersion diagram for the double nanobeam')
ax.set_xlabel("$k_x a / \\pi$")
ax.set_ylabel("$\\nu$ (THz)")
ax.set_xlim(xlims)
ax.set_ylim(ylims)
ax.legend(frameon=True)
```

```python
fig.savefig("band_diagram_double_nanobeam.pdf")
```
