## First variation: Different dephases for one sinusoidal block

We will calculate 4 band structures for different dephases: [20ยบ, 45ยบ, 90ยบ and 180ยบ]

```python
plot = False
simulate = False
save = True
```

```python
# Some parameters to describe the geometry:
index = 1.9935  # refractive index of Si3N4
ax = 3.7   #  e-7 meters
ay = 15
az = 15

# The cell dimensions
sx = 3.7 / ax
sy = 15     #size of cell in y direction (perpendicular to wvg.)
sz = 15     # size of cell in z direction (perpendicular to wvg.)


cell = mp.Vector3(sx, sy, sz)
geometry_lattice = mp.Lattice(size=mp.Vector3(sx,sy, sz))


resolution = 8
```

```python
def eps_func (Vector3):
    array3 = np.array(Vector3)
    x = array3 [0]
    y = array3 [1]
    z = array3 [2]
    

    dielec_const = 1.9935 # this can be modified
    
    dephase = 180 * np.pi / 180
    period = 3.7 / ax
    amp = 1.4 /ax
    width = 2.88 /ax
    thickness = 2 /ax
    gap = 2.2 /ax
    
    if y >= 0.5*gap and amp*np.cos(2*np.pi*(x) + dephase) + 0.5*gap+width>= y and 0.5 *thickness > np.abs(z):
        eps = mp.Medium(index = dielec_const) 
        
    elif y <= -0.5*width and -amp*np.cos(2*np.pi*(x)) - 0.5*gap-width <= y and 0.5 *thickness > np.abs(z):
        eps = mp.Medium(index = dielec_const) 
    else:
        eps = mp.Medium(epsilon = 1)
        
    if  width-amp+0.5*gap  >= np.abs(y) >= 0.5*gap and 0.5 *thickness > np.abs(z) :
        eps = mp.Medium(index = dielec_const) 

    return eps
```

```python
default_material = eps_func
```

```python
if plot == True:
    from mayavi import mlab
    sim = mp.Simulation(resolution=resolution,
                    cell_size=mp.Vector3(sx,sy,sz),
                    material_function=default_material)

    sim.init_sim()

    eps_data = sim.get_epsilon()
    eps_data = np.concatenate([eps_data,eps_data,eps_data])
    s = mlab.contour3d(eps_data, colormap="YlGnBu")
    mlab.show()
```

```python
num_bands = 10

num_k = 1
k_min = 0
k_max = 0.5 

k_points = mp.interpolate(num_k, [mp.Vector3(k_min), mp.Vector3(k_max)])
```

```python
if plot == True:    
    ms = mpb.ModeSolver(geometry_lattice=geometry_lattice,
                k_points=[mp.Vector3(0.5,0.0)],
                default_material=default_material,
                resolution=resolution,
                num_bands=num_bands);
    ms.run();
```

```python
if plot == True: 
    eps = ms.get_epsilon()
    md = mpb.MPBData(rectify=True, periods=3, resolution=resolution)
    converted_eps = md.convert(eps).T
    print(np.shape(converted_eps))
```

```python
if plot == True: 
    plt.imshow(converted_eps[180,:120,:], interpolation='spline36', cmap='binary')
    plt.axis('off')
    plt.show()
```

Now we create different band diagrams for the different dephases. We load them up and compare them.

```python
import os
path = os.path.abspath(os.getcwd())
print(path)
```

```python
f1 = np.genfromtxt(path+"/APCW_mpb_bands_dephase_45.dat", delimiter=",")[:,1:]
f2 = np.genfromtxt(path+"/APCW_mpb_bands_dephase_90.dat", delimiter=",")[:,1:]
f3 = np.genfromtxt(path+"/APCW_mpb_bands_dephase_135.dat", delimiter=",")[:,1:]
f4 = np.genfromtxt(path+"/APCW_mpb_bands_dephase_180.dat", delimiter=",")[:,1:]
```

```python
files  = [f1, f2, f3, f4]
```

```python
from scipy.interpolate import make_interp_spline, BSpline

fig, axs = plt.subplots(2,2,figsize=(18,18))
dephases  =  [45,90,135,180]

light_speed = 3e8
a = 3.7 * 1e-7

j  = 0

for axi in axs:
    for ax in axi: 
    
        ax.set_title('Dispersion diagram for the APCW. Dephase: '+str(dephases[j])+'ยบ')
        ax.set_xlabel("Wave vector $k_x$")
        ax.set_ylabel("Frequency $\\nu$ (THz)")
    
        f = files[j]
    
        j += 1

        for i in range(1,10):
            xnew = np.linspace((f[1:,1]*2).min(), (f[1:,1]*2).max(), 300) 
            spl1 = make_interp_spline(f[1:,1]*2, f[1:,-i]*light_speed/(a*10**12), k=3)  # type: BSpline
            mode_smooth = spl1(xnew)
            if i ==1:
                ax.plot(xnew, mode_smooth, c='black', linestyle='dashed', label='Dispersion modes') 
            else:
                ax.plot(xnew, mode_smooth, c='black', linestyle='dashed')

        ax.plot(np.linspace(0,1,100),np.linspace(0,1,100)*light_speed/(2*a*10**12), c='r', label ='Light line')
        ax.fill_between(np.linspace(0,1,100), np.linspace(0,1,100)*light_speed/(2*a*10**12), 1*np.ones(100)*light_speed/(2*a*10**12), color='purple', alpha=0.8, label='Light cone') 

        ax.set_ylim([320,400])
        ax.set_xlim([0.87,1])
        ax.legend(loc=2)
        
```

```python
if save == True:
    fig.savefig('Small_dephase_symmetry_breaking_bands.pdf')
```
